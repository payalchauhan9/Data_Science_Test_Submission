## Task 7 - Statistics

#### Task 7.1 (2 points):

Which type of t-test should be used to determine if there is a statistical difference between the number of strikes (`STR_total`) when there was\wasn't a warning (`WARNED=Yes\No`)

```python
print("In this case we use two tailed t-test.")
```

#### Task 7.2

Apply the test you suggested in the previous task.

```python
yes = data4[data4['WARNED'] == 'Yes']
no = data4[data4['WARNED'] == 'No']
yes_value = yes['total_struck']
no_value = no['total_struck']

from scipy import stats

null_hypothesis = "There is no statistical difference between the number of strikes when there was\wasn't a warning  "
alternate_hypothesis = "There is a statistical difference between the number of strikes when there was\wasn't a warning  "

alpha = 0.05
t_value,p_value=stats.ttest_ind(yes_value,no_value)
m = np.mean(yes_value - no_value)
s = np.std(yes_value - no_value, ddof=1)
n = len(no_value)
t_statistic_manual = m / (s / np.sqrt(n))

if p_value < alpha:
  decision = "Reject"
else:
    decision = "Fail to reject"
print("T-statistic (from scipy):", t_value)
print("P-value (from scipy):", p_value)
print("T-statistic (calculated manually):", t_statistic_manual)
print(f"Decision: {decision} the null hypothesis at alpha={alpha}.")
```

#### Task 7.3  

What do the results tell us?
Briefly interpret the results.

```python
print("Since the p-value which is 1.4845150549504365e-41 is less than the value of alpha = 0.05. So here we reject the null hypothesis. Which means that there is a statistical difference between the number of strikes when there was\wasn't a warning.")
```
